# Load images and perform some action

```python
import cv2
import glob
```

```python
path = "Images/*.*"
img_number = 1 #start an iterator for image number

for file in glob.glob(path):
    print(file)
    a = cv2.imread(file)
    #print(a)  #print arrays for each file
    c = cv2.cvtColor(a,cv2.COLOR_BGR2RGB)
    cv2.imwrite("Images/"+str(img_number)+".jpg",c)
    img_number+=1
    cv2.imshow("Color image",c)
    cv2.waitKey(1000)
    cv2.destroyAllWindows()
```

```python
#Change directory
import cv2
import glob

path = "Images/*.*"
img_number = 1  # Start an iterator for image number

for file in glob.glob(path):
    print(file)
    a = cv2.imread(file)
    c = cv2.cvtColor(a, cv2.COLOR_BGR2RGB)
    cv2.imwrite("Change/Change" + str(img_number) + ".jpg", c)  # Change destination directory here
    img_number += 1
    cv2.imshow("Color image", c)
    cv2.waitKey(1000)
    cv2.destroyAllWindows()
```

## os.path.join()

```python
import os
print(os.walk("."))
```

```python
for root,dirs,files in os.walk("."):
    print(root)
    path = root.split(os.sep)
    print(path)
    print(files)
    
```

```python
print((len(path) - 1) * '---', os.path.basename(root))  
for file in files: 
    print(len(path) * '---', file)
```
