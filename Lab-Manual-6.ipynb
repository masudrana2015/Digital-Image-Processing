# Denoising using median Filtering

```python
import cv2
from skimage.filters import median
from skimage.morphology import disk
```

```python
# img_gaussian_noice = cv2.imread("Images/noise.jpeg",0)
img_salt_pepper_noise = cv2.imread("Images/salt_pepper.jpeg",0)
img = img_salt_pepper_noise
```

```python
# cv2.imshow(img_gaussian_noice)
# cv2.imshow(img_salt_pepper_noise)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
median_using_cv2 = cv2.medianBlur(img,3)
```

```python
median_using_skimage = median(img, disk(3),mode='constant',cval=0.0)
```

```python
cv2.imshow('Orignal',img)
cv2.imshow("cv2 median", median_using_cv2)
cv2.imshow("Using Skimage",median_using_skimage)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Denoising using NLM filtering:


```python
import cv2
import numpy as np
from skimage import io,img_as_float
from skimage.restoration import denoise_nl_means,estimate_sigma
```

```python
img_gaussian_noice = img_as_float(io.imread('Images/noise.jpeg',0))
img_salt_pepper_noise=img_as_float(io.imread('Images/salt_pepper.jpeg',0))
img = img_gaussian_noice
```

```python
sigma_est = np.mean(estimate_sigma(img,channel_axis=True))
```

```python
denoise_img = denoise_nl_means(img,h=1.15*sigma_est,fast_mode=True, patch_size=5,patch_distance=3,channel_axis=None)
```

```python
cv2.imshow("Original Image",img)
cv2.imshow("NML Filter",denoise_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Denoising Works using Total Variance filtering


```python
import cv2
from skimage import io,img_as_float
from skimage.restoration import denoise_tv_chambolle
from matplotlib import pyplot as plt
```

```python
img = img_as_float(io.imread('Images/noise.jpeg',0))
plt.hist(img.flat,bins=100,range=(0,1))
denoise_img = denoise_tv_chambolle(img, weight=0.1, eps=0.0002, channel_axis=False)
plt.hist(denoise_img.flat,bins=100,range=(0,1))
cv2.imshow("Original",img)
cv2.imshow("TV Filtered",denoise_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt

# Read the images
image1 = cv2.imread('Images/noise.jpeg')
image2 = cv2.imread('Images/salt_pepper.jpeg')

# Apply Gaussian Blur to the first image
gaussian_blur_image1 = cv2.GaussianBlur(image1, (5, 5), 0)

# Apply Median Blur to the second image
median_blur_image2 = cv2.medianBlur(image2, 5)

# Plot the original and filtered images
plt.figure(figsize=(10, 8))

plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
plt.title('Original Image 1')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(cv2.cvtColor(gaussian_blur_image1, cv2.COLOR_BGR2RGB))
plt.title('Gaussian Blur')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
plt.title('Original Image 2')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(cv2.cvtColor(median_blur_image2, cv2.COLOR_BGR2RGB))
plt.title('Median Blur')

plt.show()
```
