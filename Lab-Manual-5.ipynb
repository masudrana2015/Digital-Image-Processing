## Basic image processing using opencv

```python
import cv2
import numpy as np
img=cv2.imread('Images/gojo.jpg',1)
resized=cv2.resize(img,None,fx=2,fy=2,interpolation=cv2.INTER_CUBIC)
#plt.imshow(img)


height, width = img.shape[:2]
res = cv2.resize(img,(2*width, 2*height), interpolation = cv2.INTER_CUBIC)

cv2.imshow("Original Image",img)
cv2.imshow("Resized Image", resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
img1=cv2.imread('Images/flower1.jpg')
img2=cv2.imread('Images/flower2.jpg')
img2=cv2.resize(img2,(img1.shape[1],img1.shape[0]))

merged_img=cv2.addWeighted(img1,0.5,img2,0.5,0)

cv2.imshow("Merge Image",merged_img)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

## Edge detection

## Image filtering in python - Unsharp mask

```python
from skimage import io,img_as_float
from skimage.filters import unsharp_mask
from skimage.filters import gaussian
from matplotlib import pyplot as plt

img=img_as_float(io.imread('Images/flower1.jpg',as_gray=True))
gaussian_image=gaussian(img,sigma=2,mode='constant',cval=0.0)
img2=(img-gaussian_image)*2
img3=img+img2
plt.imshow(img3,cmap='gray')
```

## skimage.filters.unsharp_mask used directly to apply the unsharp mask

```python
from skimage.filters import unsharp_mask
from skimage import io
import matplotlib.pyplot as plt

img=io.imread('Images/flower2.jpg')
unsharp_img=unsharp_mask(img,radius=3,amount=2)
fig=plt.figure(figsize=(12,12))

ax1=fig.add_subplot(1,2,1)
ax1.imshow(img,cmap='gray')
ax1.title.set_text('Input Image')
ax2=fig.add_subplot(1,2,2)
ax2.imshow(unsharp_img,cmap='gray')
ax2.title.set_text('Unsharp Image')
plt.show()
```

### Application of Gaussian blur to an image using both OpenCV (cv2.GaussianBlur) and scikit-image (skimage.filters.gaussian).


```python
import cv2
from skimage import io,img_as_float
from skimage.filters import gaussian

img_gaussian_noise=img_as_float(io.imread('Images/noise.jpeg',as_gray=True))
img_salt_peper_noise=img_as_float(io.imread('Images/salt_pepper.jpeg',as_gray=True))
img=img_salt_peper_noise

gaussian_using_cv2=cv2.GaussianBlur(img,(3,3),0,borderType=cv2.BORDER_CONSTANT)
gaussian_using_skimage=gaussian(img,sigma=1,mode='constant',cval=0.0)

cv2.imshow('Original',img)
cv2.imshow('Usung Cv2 Gaussian',gaussian_using_cv2)
cv2.imshow('using skimage',gaussian_using_skimage)
#cv2.imshow('Using Salt Paper Image',img_salt_peper_noise)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
from skimage import io, img_as_float
from skimage.filters import gaussian

img_gaussian_noise = img_as_float(io.imread('Images/noise.jpeg', as_gray=True))
img_salt_pepper_noise= img_as_float(io.imread('Images/salt_pepper.jpeg', as_gray=True))
img = img_gaussian_noise

gaussian_using_cv2 = cv2.GaussianBlur(img, (3,3), 0, borderType=cv2.BORDER_CONSTANT)
gaussian_using_skimage = gaussian(img, sigma=1, mode='constant', cval=0.0)
#sigma defines the std dev of the gaussian kernel. Slightly different than 
#how we define in cv2
cv2.imshow("Original", img)
cv2.imshow("Using cv2 gaussian", gaussian_using_cv2)
cv2.imshow("Using skimage", gaussian_using_skimage)
cv2.imshow("Using Salt Pepper Image", img_salt_pepper_noise)
cv2.waitKey(0)          
cv2.destroyAllWindows() 
```
